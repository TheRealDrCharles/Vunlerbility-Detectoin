```python
# -*- coding: utf-8 -*-
"""
The input of this file are the extracted ASTs (serialized) outputed by CodeSensor.
The output of this file are the textual sequence of ASTs.
"""

import os
import time

Project_Name = "/LibPNG"

Vulnerble = '/Vulnerable_functions'

Non_Vulnerble = '/Non_vulnerable_functions'

Working_directory = "/Users/charleswayne/Downloads/preprocessing_data"

FILE_PATH = Working_directory + Project_Name + Non_Vulnerble + '/ASTs/'

Processed_FILE = Working_directory + Project_Name + Non_Vulnerble + '/processed/'

def processing(s):
    s = s.split()
    new_str = ''
    for x in s:
        x = ''.join(x)
        new_str += x
        new_str += '_'
    return new_str[:-1]
def DepthFirstExtractASTs(file_to_process, file_name):
    lines = []
    subLines = ''
    f = open(file_to_process)
    try:
        original_lines = f.readlines()
        lines.append(file_name) # The first element is the file name.
        for line in original_lines:
            if not line.isspace(): # Remove the empty line.
                line = line.strip('\n')
                str_lines = line.split('\t')   
                if str_lines[0] != "water": # Remove lines starting with water.
                    if str_lines[0] == 'LEAF_NODE':
                        res = processing(str_lines[4])
                        lines.append(res)                            
                    if str_lines[0] == 'CALLEE': 
                        lines.append('callee')
                    if str_lines[0] == 'FUNCTION_NAME':
                        lines.append("function_name")
                    if str_lines[0] == 'ARGUMENT': 
                        lines.append('argument')
                    if str_lines[0] == 'ITERATION': 
                        lines.append('iteration')
                    if str_lines[0] == "PARAMETER_DECL":
                        lines.append('parameter_decl')     
                    if str_lines[0] == "VAR_DECL":
                        lines.append('car_decl')     
                    if str_lines[0] == 'EXPR':
                        lines.append('expr')   
                    if str_lines[0] == 'FUNCTION_CALL':
                        lines.append('function_call')
                    if str_lines[0] == 'ASSIGN_OP':
                        lines.append('assign_op')
                   
                    
                    
        subLines = ','.join(lines)
        subLines = subLines + "," + "\n"
    finally:
        f.close()
        return subLines
 
    
big_line = []
total_processed = 0

for fpathe,dirs,fs in os.walk(FILE_PATH):
    for f in fs:
        if os.path.splitext(f)[1]=='.txt':
            
            file_path = os.path.join(fpathe,f)
            res = DepthFirstExtractASTs(FILE_PATH + f, f)
            big_line.append(res)
            total_processed = total_processed + 1
            if res == '':
                print('the name is')
                print(f)
            print('none') if res == '' else print(res)
print ("Totally, there are " + str(total_processed) + " files.")
strResult = ''.join(big_line)
f1 = open(Processed_FILE + "/_proASTResult_DFT1.txt", "w")
f1.write(strResult)
f1.close()
```
